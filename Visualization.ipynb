```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
df=pd.read_csv("datafinal.csv")
df
```

```python
df.drop("Unnamed: 0",axis=1,inplace=True)
```

# Uni Variant

### Numerical

```python
sns.distplot(df["Price"],fit=norm,bins=25)
plt.show()
```

```python
#Using dist Plot()
#The price range of Normal Distribution above is typically Betweeen 0 and 50,000 with Maximum price of 2,60,000
```

```python
sns.kdeplot(df["Price"])
```

```python
#Using the kdeplot()
#It is same as previous Distribution curve where the maximum Price is around 2,60,00 and Normal Distribution curve is in range of 0 to 50,000
```

```python
plt.figure(figsize=(16,10))
sns.distplot(df["Rating"],fit=norm)
```

```python
#using dist plot()
#In the above Analysis we can observe the normal distribution curve is at its maximum in the range of  3, 4, 5  and less in one and 2 star Ratings
```

```python
plt.hist(df["Cancellation Charges"])
```

```python
#Using Hist()
#As per the Data the Cancellation charges are higher for 0 to 5,000 range as compared to other ranges
```
