### Scrapping

```python
from selenium import webdriver
```

```python
import re
import numpy as np
import pandas as pd
```

```python
location=[]
price=[]
rating=[]
by=[]
reserve=[]
can=[]
duration=[]
booking=[]
for i in range(1,60,30):
    url=f"https://www.tripadvisor.in/Attractions-g293860-Activities-c61-t220-oa{i}-India.html"
    driver=webdriver.Chrome(r"C:\Users\ADMIN\Downloads\chromedriver_win32\chromedriver.exe")
    driver.get(url)
    soup = BeautifulSoup(driver.page_source,'html.parser')
    driver.close()
    for i in soup.find_all("div",class_="XfVdV o AIbhI"):
        text=i.text
        regex=re.findall("\d+.\s(.+)",text)
        location.append(regex[0])
    for i in soup.find_all("div",class_="biGQs _P fiohW avBIb ngXxk"):
        text=i.text
        regex=re.findall("(\d+.\d+)",text)
        if regex:
             price.append(regex[0])
        else:
             price.append(np.nan)
    for i in soup.find_all("div",class_="jVDab o W f u w JqMhy"):
        text=i.text
        regex=re.findall("\d+",text)
        if regex:
             rating.append(regex[0])
        else:
             rating.append(np.nan)
    for i in soup.find_all("div",class_="NxKBB"):
        text=i.text
        regex=re.findall("\dBy(.*)\s",text)
        if regex:
            by.append(regex[0])
    for i in soup.find_all("article",class_="GTuVU XJlaI rHoxO"):
        text=i.text
        regex=re.findall("(?:adult|group)Reserve",text)
        reserve.append(regex[0])
    for i in soup.find_all("div",class_="FUUWF P"):
        text=i.text
        regex=re.findall("^(F.*)N",text)
        if regex:
            can.append(regex[0])
    for i in soup.find_all("div",class_="bRMrl _Y K fOSqw"):
        text=i.text
        regex=re.findall(".*",text)
        duration.append(regex[0])
    for i in soup.find_all("a",class_="rmyCe _G B- z _S c Wc wSSLS jWkoZ XDcOZ"):
        text=i.text
        regex=re.findall(".*",text)
        booking.append(regex[0])
    print("*******************************************************************************")
print(by)
print(len(by))
```

```python
print(len(location))
print(len(price))
print(len(rating))
print(len(by))
print(len(reserve))
print(len(can))
print(len(duration))
print(len(booking))
```

```python
a=pd.DataFrame(location,columns=["Location"])
b=pd.DataFrame(price,columns=["Price"])
c=pd.DataFrame(rating,columns=["rating"])
d=pd.DataFrame(by,columns=["Agency/Company"])
e=pd.DataFrame(reserve,columns=["Reserve Type"])
f=pd.DataFrame(can,columns=["Availability of Cancellation"])
g=pd.DataFrame(duration,columns=["Tour Duration"])
h=pd.DataFrame(booking,columns=["Booking Availability"])
```

```python
final=pd.concat([a,b,c,d,e,f,g,h],axis=1)
```
